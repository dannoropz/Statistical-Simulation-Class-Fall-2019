Hagan un programa (en excel o en R), que permita simular el proceso de obtener la primera y segunda muestra considerando como parámetros el tamaño N de la población de interés, el tamaño de la primera y segunda muestra y como datos a estimar son: de qué tamaño debe ser n1 y n2para obtener una buena aproximación y ver cómo se afecta por el tamaño N.

```python
def mark_recapture(n1, n2, N):
    true_prop = n1/N # proporción real de animales marcados en toda la población
    marked = np.ones(n1) #cada animal marcado será un 1
    not_marked = np.zeros(N-n1) #cada animal de la población que no fue marcado es un 0
    population = np.concatenate((marked, not_marked)) #la población total está hecha de animales marcados y no marcados
    v2_sample = np.random.choice(population, size=n2, replace=False) #tomo una muestra de tamaño n2 y cada animal se asume 
                                                                # que tiene la misma prob. de salir, esté marcado o no
    marked_v2 = np.sum(v2_sample)   #la suma de los 1 que obtuvimos en n2 son los animales marcados que obtuvimos 
    estimated_prop = marked_v2 / n2 #en este estimador se asume que k/K = n/N, o sea, las proporciones entre muestra y 
                                    #población son iguales
    return true_prop, estimated_prop
```

Observa cómo cambia qué tan bueno es el estimador según cambia el tamaño de las muestras, para un tamaño de N fijo:

```python
for i in range(len(s_sizes)):
    s_size = s_sizes[i]
    tp, ep = mark_recapture(n1,s_size,N)
    tps.append(tp)
    eps.append(ep)
```

Grafiquen una ejemplo del proceso considerando el intervalo de tiempo [0,100]

```python
def poi_nohomo(horrible_function,n):
    lmbda = 5 ## we never have values above 5 for the function above
    TT = np.random.exponential(scale=1/lmbda, size=n)
    s = np.cumsum(TT)
    u = np.random.uniform(size=n)
    horrible = [horrible_function(s[i])/lmbda for i in range(n)] # vector of values with which I compare each u
    ss = s[u <= horrible] 
    Ns = range(len(ss))
    fig = plt.figure(figsize=(8,6))
    plt.step(ss,Ns, color='red')
```

```python
poi_nohomo(horrible_function, 500)
plt.title('Gráfica del proceso de Poisson no homogéneo')
```

### 8. Comenzando a mediodía, los comensales llegan a un restaurante de acuerdo a unproceso Poisson a una tasa de 5 clientes por minuto. El tiempo que cada clientepasa comiendo en el restaurante tiene una distribución exponencial con media de 40minutos, independiente de otros clientes e independiente de los tiempos de arribo.Encuentra la distribución así como la media y varianza, del número de comensalesen el restaurante a las 2:00pm. Simular el restaurante para verificar los resultadosobtenidos.

```python
def simulated_restaurant(lmbda, n):
    # proceso de Poisson de ingreso
    TA = np.random.exponential(scale=1/lmbda, size=n)
    tau = np.cumsum(TA)
    Ns = range(len(tau))
    
    # proceso de salidas con tiempo exponencial
    eating_times = np.random.exponential(scale=40, size=len(tau)) #genero cuánto tiempo se tarda en comer cada comensal
    exit_times = [tau[i] + eating_times[i] for i in range(len(tau))] #genero los tiempos de salida
    exit_times_s = np.sort(exit_times) #los ordeno de menor a mayor
    Ne = range(len(exit_times))
    
    return tau, Ns, exit_times_s, Ne
```

```python
tau, Ns, exit_times_s, Ne = simulated_restaurant(5, 700)
```

Abajo estoy ilustrando el proceso de ingreso de clientes (en azul) y el de salida de clientes (en naanja), entonces para el minuto 120 (que es dos horas después de las 12:00, o sea a las 2:00 pm) la diferencia entre la línea azul y la línea anaranjada son la cantidad de comensales en el restaurante. ### 10. Una cadena de Markov de nacimiento y muerte es un proceso estocástico con un espacio de estados numerable infinito y dos tipos de transiciones: nacimientos de i a i+1 y muertes de i a i−1. 

Ver pdf adjunto para la descripción de la cadena.

```python
def create_birth_and_death(N, initial=10):
    U = np.random.uniform(size=(N,3)) # matriz de N x 3 de aleatorios uniformes. 
    U_sum = np.sum(U, axis=1) # vector de N x 1 donde cada celda es la suma de los tres uniformes del renglón i de U
    probs = [U[i]/U_sum[i] for i in range(len(U))] ## obtengo las probabilidades para cada uno de los N estados i
    BD = [initial] # establezco el estado inicial del proceso
    transition_states = np.array([initial-1,initial,initial+1]) #establezco las transiciones que se irán actualizando
    for i in range(1,N):
        BD.append(np.random.choice(transition_states, p=probs[i])) #esto va generando la lista con los valores del proceso
        transition_states = np.array([BD[i]-1, BD[i], BD[i]+1]) #esto actualiza los estados  
    return BD
```

Aquí una visualización: 

```python
BD = create_birth_and_death(100, 5)
fig = plt.figure(figsize=(15,6))
plt.plot(range(100), BD)
```

Ahora vamos a simularlo 500 veces.

```python
BDs = []
```

```python
for i in range(500):
    BD = create_birth_and_death(500,10)
    BDs.append(BD)
```
